```python
import pandas as pd

df = pd.read_csv("Univeristies.csv")
df.head(15)
```

```python
group = df.groupby("InstitutionName").agg({"IN_STATE_TUIT_FEE": "mean"})
university = list(group.index)
university
```

```python
values = {}

for i in university:
    val = []
    for index, row in df.iterrows():
        if row['InstitutionName'] == i:
            val.append(row['IN_STATE_TUIT_FEE'])
    print(val)
    values[i] = val
    
            
```

```python
values
```

```python
dfNew = pd.DataFrame(values)
dfNew.head()
```

```python
dfNew.to_csv("new_College.csv")
```

```python
def read_data():
    
    import pandas as pd
    df_university = pd.read_csv("new_College.csv")
    df_university.head()
    columnName1 = list(df_university.columns)
    dictt2 = {}
    for i in columnName1:
        dictt2[i] = list(df_university[i])   
    df_standerd = pd.read_csv("median_level.csv")
    df_standerd = df_standerd.dropna()
    df_standerd.head()

    columnName = list(df_standerd.columns)
    
    dictt = {}
    for i in columnName:
        dictt[i] = list(df_standerd[i])   
    finaldict = { 'standerd' : dictt, 'college': dictt2 }

    return finaldict
```
